```python
from matplotlib import pyplot as plt
```

```python
x=[1,2,3]
x
```

```python
y=[4,5,6]
y
```

```python
plt.plot(x,y)
plt.title('Info')  #Title of the plot
plt.ylabel('Y axis') #Label of x-axis.
plt.xlabel('X axis') #Label of y-axis.
plt.show()
```
