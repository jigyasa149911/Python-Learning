```python
from matplotlib import pyplot as plt
import numpy as np
# Subplot() function can be used to plot two or more plots in one figure.
```

```python
t1=np.arange(0.0,20.0)
t2=np.arange(0.0,20.0)
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
plt.subplot(221)
plt.plot(t1,f(t1),t2,f(t2))
plt.subplot(222)

plt.title('subplot(211)')
plt.show()
```
