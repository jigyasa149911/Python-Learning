```python
from matplotlib import pyplot as plt
```

```python
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],label='BMW',width=0.2) # plotting the bar graphs by giving coordinates.
plt.bar([0.75,1.75,2.75,3.75,4.75],[30,50,90,60,80],label='Audi',width=0.2,color='r')  # plotting the bar graphs by giving coordinates.
plt.xlabel('Days') # Giving label to X-axis.
plt.ylabel('Distance(kms)') #Giving label to Y-axis.
plt.title('Information') # Giving title to the Bar graph.
plt.show()
```
