```python
from matplotlib import pyplot as plt
# Scatter plot is a 2-D data visualisation that uses dots to represent the values obtained for two different variables.
```

```python
x=[1,1.5,2,2.5,3,3.5,3.6]
y=[7.5,8,8.5,9,9.5,10,10.5]
x1=[8,8.5,9,9.5,10,10.5,11]
y1=[3,3.5,3.7,4,4.5,5,5.2]
```

```python
plt.scatter(x,y) #Makes scatter plot by given coordinate.
plt.scatter(x1,y1)
plt.xlabel('saving*100')
plt.ylabel('income*1000')
plt.title('Scatter plot')
plt.show()
```
