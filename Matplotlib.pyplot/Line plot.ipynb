```python
from pylab import *
```

```python
t=arange(0.0,2.0,0.01)
s=sin(2.5*pi*t)
plot(t,s)
xlabel('time(s)') #gives label to X-axis.
ylabel('voltage(mV)')  #gives label to Y-axis.
title('sine wave')
grid(True) # Turns on the grid.
show()
```

```python
t=arange(0.0,2.0,0.01)
s=cos(2.5*pi*t)
plot(t,s)
xlabel('time(s)') #gives label to X-axis.
ylabel('voltage(mV)')  #gives label to Y-axis.
title('cosine wave')
grid(True) # Turns on the grid.
show()
```
