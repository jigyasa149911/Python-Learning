```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
objects=('Python','C++','Java','Perl','Scala','Lisp')
n=np.arange(len(objects))
performance=[10,8,6,4,2,1]
plt.bar(n,performance)
plt.xticks(n,objects)
plt.ylabel('Usage')
plt.xlabel('Language')
plt.title('Programming language usage')
plt.show()
```

```python
# Horizontal bar graph
objects=('Python','C++','Java','Perl','Scala','Lisp')
n=np.arange(len(objects))
performance=[10,8,6,4,2,1]
plt.barh(n,performance)
plt.xticks(n,objects)
plt.ylabel('Usage')
plt.xlabel('Language')
plt.title('Programming language usage')
plt.show()
```
