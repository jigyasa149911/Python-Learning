```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Question-1:Write a python program to display a bar chart of popularity of programming languages.
# Sample Data:
# Programming Languages : Java , Python , PHP , JavaScript , C# , C++.
# Popularity : 22.2 , 17.6 , 8.8 , 8 , 7.7 , 6.7
programming_languages=('Java','Python','PHP','JavaScript','C#','C++')
n=np.arange(len(programming_languages))
popularity=[22.2,17.6,8.8,8,7.7,6.7]
plt.bar(n,popularity)
plt.xticks(n,programming_languages)
plt.xlabel('Programming Language')
plt.ylabel('Popularity')
plt.title('Popular Programming Languages')
plt.show()
```

```python
# Question 2 : Write a Python program to create a bar plot of scores by group and gender.
# Use multiple X values on same chart for men and women.
# Sample data
# Means(men):(22 , 30 , 35 , 35 , 26)
# Means(women):(25 , 32 , 30 , 35 , 29)
men=[22 , 30 , 35 , 35 , 26]
women=[25 , 32 , 30 , 35 , 29]
no_of_women=[100 , 34 , 45 , 46 , 35]
no_of_men=[120 , 78 , 54 , 100 , 16]
plt.bar(men,no_of_men)
plt.bar(women,no_of_women)
plt.xlabel('Gender')
plt.ylabel('Group')
plt.show()
```

```python
# Question-3 : Write a Python program to add textures (black and white ) to bars and wedges.
men=[22 , 30 , 35 , 35 , 26]
no_of_men=[120 , 78 , 54 , 100 , 16]

women=[25 , 32 , 30 , 35 , 29]
no_of_women=[100 , 34 , 45 , 46 , 35]

plt.bar(men,no_of_men,hatch='/',color='black')

plt.bar(women,no_of_women,hatch='-',color='white')

plt.xlabel('Gender')
plt.ylabel('Group')

plt.show()
```

```python
# Question 4 : Write a python program to create a pie chart with a title of popularity of Programming Languages.
# # Sample Data:
# Programming Languages : Java , Python , PHP , JavaScript , C# , C++.
# Popularity : 22.2 , 17.6 , 8.8 , 8 , 7.7 , 6.7
programming_languages=('Java','Python','PHP','JavaScript','C#','C++')
#n=np.arange(len(programming_languages))
popularity=[22.2,17.6,8.8,8,7.7,6.7]
explode=(0.1,0,0,0,0,0)
colors=['gold','red','blue','green','lightcoral','pink']
patches , texts = plt.pie(popularity,explode=explode,labels=programming_languages,shadow=True,startangle=90,colors=colors)
plt.legend(patches,programming_languages,loc='lower center')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
 
df = pd.DataFrame([[10, 20, 30, 40], [7, 14, 21, 28], [55, 15, 8, 12],
                    [15, 14, 1, 8]],
                  columns=['Apple', 'Orange', 'Banana', 'Pear'],
                  index=['Basket1', 'Basket2', 'Basket3', 'Basket4']
                  )
 
plt.imshow(df, cmap="cool")
plt.xticks(range(len(df)),df.columns, rotation=20)
plt.yticks(range(len(df)),df.index)
plt.show()
```
